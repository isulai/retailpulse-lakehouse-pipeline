
# RetailPulse Gold Layer

Purpose:
Create business aggregates from Silver data.

Outputs:
- gold_region_revenue
- gold_category_revenue
- gold_top_customers


### Create Gold Region Revenue

```python
from pyspark.sql.functions import col, sum

gold_region_revenue_df = (
    silver_df.groupBy("region")
    .agg(sum("revenue").alias("total_revenue"))
)

display(gold_region_revenue_df)
```


###Now Persist It As A Gold Table

```python
silver_df = spark.table("workspace.retailpulse.silver_sales_enriched")
```

```python
gold_region_revenue_df.write.format("delta") \
    .mode("overwrite") \
        .saveAsTable("workspace.retailpulse.gold_region_revenue")
```

```python
from pyspark.sql.functions import col, sum

gold_category_revenue_df = (silver_df
                            .groupBy("category").agg(sum("revenue").alias("total_revenue")
                                                     ))
display(gold_category_revenue_df)
```

```python
from pyspark.sql.functions import col, sum

gold_top_customers_df = (
    silver_df.groupBy("customer_name")
    .agg(sum("revenue").alias("total_revenue"))
    .orderBy(col("total_revenue").desc())
)

display(gold_top_customers_df)
```

```python
gold_top_customers_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.retailpulse.gold_top_customers")
```

```python
gold_category_revenue_df.write.format("delta") \
    .mode("overwrite") \
        .saveAsTable("workspace.retailpulse.gold_category_revenue")
```
