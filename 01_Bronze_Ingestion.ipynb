
# RetailPulse Bronze Layer

Purpose:
Create raw Delta tables from source data.

Tables:
- bronze_customers
- bronze_products
- bronze_sales
- bronze_returns

```python
customers_data = [
    (1, "Sulai", "South"),
    (2, "Rahul", "North"),
    (3, "Ali", "West"),
    (4, "John", "East"),
    (5, "Priya", "South")
]

customers_df = spark.createDataFrame(customers_data, ["customer_id", "customer_name", "region"])

display(customers_df)
```

```python
#Products Data

products_data = [
    (101, "Laptop", "Electronics", 50000),
    (102, "Phone", "Electronics", 30000),
    (103, "Headphones", "Accessories", 2000),
    (104, "Office Chair", "Furniture", 8000),
    (105, "Desk", "Furniture", 12000)
]


products_df = spark.createDataFrame(products_data, ["product_id", "product_name", "category", "price"])

display(products_df)
```

```python
# Sales Data

sales_data = [
    (1, 1, 101, 2, "2026-06-01"),
    (2, 2, 102, 1, "2026-06-01"),
    (3, 3, 103, 3, "2026-06-02"),
    (4, 4, 104, 1, "2026-06-02"),
    (5, 5, 105, 2, "2026-06-03")
]

sales_df = spark.createDataFrame(sales_data, ["sale_id", "customer_id", "product_id", "quantity", "sale_data"])

display(sales_df)
```

```python
# Returns Data:

returns_data = [
    (1, 2, "2026-06-05"),
    (2, 4, "2026-06-06")
]

returns_df = spark.createDataFrame(returns_data, ["return_id", "sale_id", "return_date"])

display(returns_df)
```

### From here on we are creating the Bronze Tables using the Source Datas:

```python
# Create Bronzee Customers:

customers_df.write.format("delta").mode("overwrite") \
    .saveAsTable("workspace.retailpulse.bronze_customers")

```

```python
# Create Bronze Products:

products_df.write.format("delta").mode("overwrite") \
    .saveAsTable("workspace.retailpulse.bronze_products")
```

```python
# Create Bronze Sales

sales_df.write.format("delta").mode("overwrite") \
    .saveAsTable("workspace.retailpulse.bronze_sales")
```

```python
# Create Bronze Returns

returns_df.write.format("delta").mode("overwrite") \
    .saveAsTable("workspace.retailpulse.bronze_returns")
```
